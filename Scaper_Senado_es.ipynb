```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# Función auxiliar para obtener el texto de un elemento si existe
def get_text(element):
    if element:
        return element.text
    else:
        return None
```

```python
def scrap(link): 
  url='https://www.senado.es/legis14/publicaciones/xml/senado/ds/'+link

  response = requests.get(url)
  xml_content = response.content

  # Analizar el XML con Beautiful Soup
  soup = BeautifulSoup(xml_content, "xml")

  # Extraer los datos necesarios
  apartado = soup.findAll('apartado')

  cabecera=soup.find('TITULO')
  anno=soup.find('ANNO')
  fecha=soup.find('FECHA')
  cve=soup.find('CVE')

  # Crear una lista de diccionarios con los datos extraídos
  for punto in apartado:
      if punto:
          titulo = get_text(punto.find("titulo"))
          texto = get_text(punto.find("texto"))
          punto=punto.find("punto")
          autor=get_text(punto.find("autor"))
          expediente=get_text(punto.find("expediente"))
                
          data.append({
              "Cabecera": cabecera, 
              "Año": anno, 
              "Fecha": fecha, 
              "CVE": cve, 
              "Titulo": titulo,
              "Autor": autor,
              "Texto": texto,
              "Punto": punto,
              "Expediente": expediente
              })
```

```python
# Este scrap tarda unos 5 minutos
data=[]

for x in range(1,135):
  link='DS_P_14_'+str(x)+'.XML'
  scrap(link)

# Crear un DataFrame de Pandas con los datos
df_plenos = pd.DataFrame(data)
```
